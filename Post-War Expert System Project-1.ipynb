## Define the Knowledge Base

```python
# Define the Enhanced Knowledge Base
class DamageAssessment(KnowledgeEngine):

    # Declare initial facts
    @DefFacts()
    def initial_facts(self):
        yield Fact(action="assess_damage")

    # Rule: Ghost Database Problem
    @Rule(Fact(action="assess_damage"), Fact(data_available=False))
    def ghost_database_problem(self):
        print("Warning: Ghost Database Problem - Missing records. Proceeding with caution.")

    # Rule: Adhoc Construction
    @Rule(Fact(action="assess_damage"), Fact(construction_type="adhoc"))
    def adhoc_construction(self):
        print("Warning: Adhoc construction detected. Structural integrity may be compromised.")

    # Rule: Overlapping Echoes (Duplicate Claims)
    @Rule(Fact(action="assess_damage"), Fact(compensation_claims=P(lambda x: x > 1)))
    def overlapping_echoes(self):
        print("Warning: Overlapping Echoes - Multiple compensation claims detected. Verify data consistency.")

    # Rule: Uncoordinated Assessment
    @Rule(Fact(action="assess_damage"), Fact(coordinated_assessment=False))
    def uncoordinated_assessment(self):
        print("Warning: Uncoordinated assessment. Risk of redundant or conflicting data.")

    # Rule: Broken Compass (Unqualified Engineers)
    @Rule(Fact(action="assess_damage"), Fact(engineer_qualified=False))
    def broken_compass(self):
        print("Warning: Broken Compass - Assessment conducted by unqualified personnel. Results may be unreliable.")

    # Rule: Inconsistent Reports
    @Rule(Fact(action="assess_damage"), Fact(report_discrepancy=True))
    def inconsistent_reports(self):
        print("Warning: Inconsistent reports detected. Verify damage classification.")

    # Rule: Balloon Effect (Overstated Claims)
    @Rule(Fact(action="assess_damage"), Fact(damage_overstatement=True))
    def balloon_effect(self):
        print("Warning: Balloon Effect - Damage claims may be inflated. Verify actual extent of damage.")

    # Rule: Unverified Claim
    @Rule(Fact(action="assess_damage"), Fact(claim_verification=False))
    def unverified_claim(self):
        print("Warning: Unverified claim. Risk of misinformation.")

    # Rule: Conflict Debris Hazard
    @Rule(Fact(action="assess_damage"), Fact(conflict_debris=True))
    def debris_hazard(self):
        print("Warning: Conflict debris present. Access may be hazardous.")

    # Rule: Hazardous Materials
    @Rule(Fact(action="assess_damage"), Fact(hazardous_materials=True))
    def hazardous_site(self):
        print("Warning: Hazardous materials detected. Inspection requires specialized equipment.")

    # Rule: Shifting Sands (Unstable Terrain)
    @Rule(Fact(action="assess_damage"), Fact(terrain_stability="dynamic"))
    def shifting_sands(self):
        print("Warning: Shifting Sands - Terrain instability detected. Building integrity may be compromised.")

    # Rule: Empty Toolbox (Insufficient Funding)
    @Rule(Fact(action="assess_damage"), Fact(funding_available=False))
    def empty_toolbox(self):
        print("Warning: Empty Toolbox - Insufficient funding. Assessments may be incomplete or superficial.")

    # Rule: Superficial Assessment
    @Rule(Fact(action="assess_damage"), Fact(assessment_quality="superficial"))
    def superficial_assessment(self):
        print("Warning: Superficial assessment conducted. Risk of overlooking critical damage.")
```

## Define Backward Chaining Logic

```python
# Backward Chaining Logic (Optional)
def backward_chaining(goal, facts):
    print(f"\nGoal: {goal}")

    # Define the rules manually for backward chaining
    rules = {
        "Ghost Database Problem": lambda f: f.get("data_available") is False,
        "Adhoc Construction": lambda f: f.get("construction_type") == "adhoc",
        "Overlapping Echoes": lambda f: f.get("compensation_claims", 0) > 1,
        "Uncoordinated Assessment": lambda f: f.get("coordinated_assessment") is False,
        "Broken Compass": lambda f: f.get("engineer_qualified") is False,
        "Inconsistent Reports": lambda f: f.get("report_discrepancy") is True,
        "Balloon Effect": lambda f: f.get("damage_overstatement") is True,
        "Unverified Claim": lambda f: f.get("claim_verification") is False,
        "Conflict Debris Hazard": lambda f: f.get("conflict_debris") is True,
        "Hazardous Site": lambda f: f.get("hazardous_materials") is True,
        "Shifting Sands": lambda f: f.get("terrain_stability") == "dynamic",
        "Empty Toolbox": lambda f: f.get("funding_available") is False,
        "Superficial Assessment": lambda f: f.get("assessment_quality") == "superficial",
    }

    # Check if the goal is satisfied by the facts
    if goal in rules and rules[goal](facts):
        print(f"The goal '{goal}' is satisfied.")
    else:
        print(f"The goal '{goal}' is NOT satisfied.")
```

##  Define Test Cases

```python
# Test Cases
def test_expert_system():
    engine = DamageAssessment()

    # Forward Chaining Test Cases
    print("\n--- Forward Chaining ---")

    # Case 1: Ghost Database, Adhoc Construction, Unqualified Engineer, Hazardous Materials, Insufficient Funding
    print("\nCase 1:")
    engine.reset()
    engine.declare(
        Fact(action="assess_damage"),
        Fact(data_available=False),
        Fact(construction_type="adhoc"),
        Fact(engineer_qualified=False),
        Fact(hazardous_materials=True),
        Fact(funding_available=False)
    )
    engine.run()

    # Case 2: Overlapping Echoes, Uncoordinated Assessment, Inconsistent Reports, Balloon Effect
    print("\nCase 2:")
    engine.reset()
    engine.declare(
        Fact(action="assess_damage"),
        Fact(compensation_claims=3),
        Fact(coordinated_assessment=False),
        Fact(report_discrepancy=True),
        Fact(damage_overstatement=True)
    )
    engine.run()

    # Case 3: Conflict Debris, Shifting Sands, Superficial Assessment
    print("\nCase 3:")
    engine.reset()
    engine.declare(
        Fact(action="assess_damage"),
        Fact(conflict_debris=True),
        Fact(terrain_stability="dynamic"),
        Fact(assessment_quality="superficial")
    )
    engine.run()

    # Backward Chaining Test Cases
    print("\n--- Backward Chaining ---")

    # Case 1: Check for 'Ghost Database Problem'
    facts1 = {
        "data_available": False,
        "construction_type": "adhoc",
        "engineer_qualified": False,
        "hazardous_materials": True,
        "funding_available": False
    }
    backward_chaining("Ghost Database Problem", facts1)

    # Case 2: Check for 'Balloon Effect'
    facts2 = {
        "compensation_claims": 3,
        "coordinated_assessment": False,
        "report_discrepancy": True,
        "damage_overstatement": True
    }
    backward_chaining("Balloon Effect", facts2)

    # Case 3: Check for 'Shifting Sands'
    facts3 = {
        "conflict_debris": True,
        "terrain_stability": "dynamic",
        "assessment_quality": "superficial"
    }
    backward_chaining("Shifting Sands", facts3)
```

##  Run the Test Cases

```python
# Run test cases
test_expert_system()
```
